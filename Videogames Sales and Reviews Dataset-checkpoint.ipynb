```python
plt.figure(figsize=(10, 6))
df['Genre'].value_counts().plot(kind='bar')
plt.title('Genre Distribution')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['Global'])
plt.title('Global Sales Over the Years of every product')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
platform_counts = df['Platform'].value_counts()
platform_counts.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Platform Distribution')
plt.xlabel('Platform')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Review'], bins=20, color='red', alpha=0.7)
plt.title('Distribution of Reviews')
plt.xlabel('Review Score')
plt.ylabel('Frequency')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df.groupby('Year')['Global'].sum(), marker='o', color='skyblue')
plt.title('Net Global Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.grid(True)
plt.show()
```
